```python
%pylab inline
import requests
import urllib2
import matplotlib.pyplot as plt

import matplotlib.image as mpimg # mpimg 用于读取图片
url="https://hp2.591.com.tw/house/active/2017/10/29/150925578065594607_210x158.crop.jpg"
#f= requests.get("https://hp2.591.com.tw/house/active/2017/10/29/150925578065594607_210x158.crop.jpg")
img_request = urllib2.Request(url)
img_response = urllib2.urlopen(img_request)
img = img_response.read()

f=open("/tmp/tmp.jpg","wb")
f.write(img);
f.close()
image_inst= mpimg.imread("/tmp/tmp.jpg")

plt.imshow(image_inst)
plt.show()
```

```python
#a=soup.find('iframe',attrs={'class':'myframe'})
#print a.attrs['src']
q=soup.find('div',attrs={'class','contenter banner'})
len(q)
gps_section=q.find_all('textarea',attrs={'class':'datalazyload'})
len(gps_section)
string=gps_section[0].text
import re
re_result=re.search('src="(.*?)"',string)
re_result.group(1)
```
